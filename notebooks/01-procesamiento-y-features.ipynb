```python
#Hay que estandarizar la prieba puesto que en el caso de este usuario la prueba sutvo las siguientes caracteristica
#Experimental setting 
class setting:
    """ Indica los parametros del experimento  """
    basal_tiempo: float=2*60 #min
    centro_tiempo:  float=1*60
    derecha_tiempo: float=1*60
    izquierda_tiempo: float=1*60
    adelante_tiempo: float=1*60
    atras_tiempo: float=1*60
    reposo_tiempo: float=2*60
    secuencia: list

setting.secuencia=[setting.basal_tiempo,setting.derecha_tiempo,
                  setting.centro_tiempo,setting.izquierda_tiempo,
                  setting.centro_tiempo,setting.adelante_tiempo,
                  setting.centro_tiempo,setting.atras_tiempo,
                  setting.reposo_tiempo]


#secuencia=[basal_tiempo,derecha_tiempo,centro_tiempo,
#        izquierda_tiempo,centro_tiempo,adelante_tiempo,centro_tiempo,atras_tiempo,reposo_tiempo]

init_time=0
c=0
for tiempo in setting.secuencia:
    tiempo=init_time+tiempo
    init_time=tiempo
    setting.secuencia[c]=init_time
    c+=1
print(setting.secuencia)
```

```python
def segmentar_dataframe(df, timestamps, labels, ignore_labels=None):
    """
    Segmenta un DataFrame basado en una secuencia de tiempos y etiquetas.

    Args:
        df (pd.DataFrame): El DataFrame completo con la columna 'time_seconds'.
        timestamps (list): Una lista de tiempos que marcan el FINAL de cada segmento.
        labels (list): Una lista de etiquetas de texto correspondientes a cada segmento.
        ignore_labels (list, optional): Una lista de etiquetas a ignorar. Por defecto es None.

    Returns:
        dict: Un diccionario donde las claves son las etiquetas y los valores son
              los DataFrames segmentados.
    """
    if ignore_labels is None:
        ignore_labels = []
        
    segmented_dfs = {}
    start_time = 0
    
    print("Iniciando segmentación dinámica...")
    
    # Usamos zip para iterar sobre tiempos y etiquetas al mismo tiempo
    for end_time, label in zip(timestamps, labels):
        
        # Si la etiqueta actual está en la lista de ignorados, la saltamos
        if label in ignore_labels:
            print(f"  - Ignorando segmento '{label}' (hasta {end_time}s)")
            start_time = end_time # Actualizamos el tiempo de inicio para el siguiente ciclo
            continue
            
        print(f"  - Segmentando '{label}' desde {start_time:.2f}s hasta {end_time:.2f}s")
        
        # Filtramos el DataFrame para el rango de tiempo actual
        segment = df[(df['time_seconds'] > start_time) & (df['time_seconds'] <= end_time)].copy()
        
        # Guardamos el segmento en nuestro diccionario
        segmented_dfs[label] = segment
        
        # El tiempo final de este segmento es el tiempo de inicio del siguiente
        start_time = end_time
        
    print("\n¡Segmentación dinámica completada!")
    return segmented_dfs
```

```python

labels=["basal","derecha","centro","izquierda","centro","adelante","centro","atras","reposo"]

segmented_dfs=segmentar_dataframe(df,setting.secuencia,labels,ignore_labels="centro")


print("\nVerificando los tamaños de cada segmento generado:")
for clase, segment_df in segmented_dfs.items():
    print(f"  - Muestras en '{clase}': {segment_df.shape[0]}")
```

```python
# --- Guardar los Nuevos DataFrames Segmentados con un Identificador ---

# 1. Define un identificador para este nuevo conjunto de datos
identificador_nuevo = select_sujeto # O el nombre que prefieras

# 2. Define la ruta a la carpeta 'processed'
path_processed = '../data/processed/'

print(f"Guardando {len(segmented_dfs)} archivos del sujeto '{identificador_nuevo}' en: {path_processed}")

# 3. Itera sobre el diccionario y guarda cada DataFrame
for clase, df_segmento in segmented_dfs.items():
    # Construimos el nombre del archivo dinámicamente
    nombre_archivo = f"{clase}_data_{identificador_nuevo}.parquet"
    ruta_completa = path_processed + nombre_archivo
    
    # Guardamos el archivo
    df_segmento.to_parquet(ruta_completa)
    print(f"  - Archivo guardado: {nombre_archivo}")

print("\n¡Nuevos archivos guardados exitosamente!")
```
