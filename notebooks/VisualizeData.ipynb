```python
file_path1 = '/content/drive/My Drive/Colab Notebooks/Hernando/Datos/Datos_user_1.txt'
file_path2 = '/content/drive/My Drive/Colab Notebooks/Hernando/Datos/Datos_user_2.txt'
file_path3 = '/content/drive/My Drive/Colab Notebooks/Hernando/Datos/Datos_user_3.txt'

# Use a space as the separator, skip the first 4 metadata rows.
df1 = pd.read_csv(file_path1, sep=',\s+', engine='python', skiprows=4)
df2 = pd.read_csv(file_path2, sep=',\s+', engine='python', skiprows=4)
df3 = pd.read_csv(file_path3, sep=',\s+', engine='python', skiprows=4)

# Your original code to select columns
#df1 = df1.iloc[:, 0:25]

# To display the first few rows of the DataFrame
display(df1.head())
display(df2.head())
display(df3.head())
```

```python
df1['Elapsed Time'] = df1['Timestamp'] - df1['Timestamp'].iloc[0]
df2['Elapsed Time'] = df2['Timestamp'] - df2['Timestamp'].iloc[0]
df3['Elapsed Time'] = df3['Timestamp'] - df3['Timestamp'].iloc[0]
display(df1.head())
display(df2.head())
display(df3.head())
```

```python
exg_cols = [f'EXG Channel {i}' for i in range(8)]
df1_c = df1[(df1[exg_cols] != 0).any(axis=1)].copy()
df2_c = df2[(df2[exg_cols] != 0).any(axis=1)].copy()
df3_c = df3[(df3[exg_cols] != 0).any(axis=1)].copy()
display(df1_c.head())
display(df2_c.head())
display(df3_c.head())
```

```python
fig, ax = plt.subplots(figsize=(60, 5))
ax.set_title("Timestamp vs EXG Channel 3 for users 1 & 2")
ax.plot(df1_c['Elapsed Time'], df1_c['EXG Channel 3'], label='EXG Channel 3 User 1')
ax.plot(df2_c['Elapsed Time'], df2_c['EXG Channel 3'], label='EXG Channel 3 User 2')
ax.plot(df3_c['Elapsed Time'], df3_c['EXG Channel 3'], label='EXG Channel 3 User 3')
ax.set_xlabel("Elapsed Time")
ax.set_ylabel("EXG Value 3 for users 1, 2 & 3")
ax.legend()
```

```python
fig, ax = plt.subplots(figsize=(60, 5))
ax.set_title("Timestamp vs EXG Channel 4 for users 1 & 2")
ax.plot(df1_c['Elapsed Time'], df1_c['EXG Channel 4'], label='EXG Channel 4 User 1')
ax.plot(df2_c['Elapsed Time'], df2_c['EXG Channel 4'], label='EXG Channel 4 User 2')
ax.plot(df3_c['Elapsed Time'], df3_c['EXG Channel 4'], label='EXG Channel 3 User 3')
ax.set_xlabel("Elapsed Time")
ax.set_ylabel("EXG Value 4 for users 1, 2 & 3")
ax.legend()
```
