```python
file_path = '/content/drive/My Drive/Colab Notebooks/Hernando/Datos/Datos_user_1.txt'

# Use a space as the separator, skip the first 4 metadata rows.
df1 = pd.read_csv(file_path, sep=',\s+', engine='python', skiprows=4)

# Your original code to select columns
df1 = df1.iloc[:, 0:25]

# To display the first few rows of the DataFrame
display(df1.head())
```

```python
fig, ax = plt.subplots(figsize=(60, 5))
ax.set_title("Timestamp vs EXG Channels 3 and 4")
ax.plot(df1['Timestamp'], df1['EXG Channel 3'], label='EXG Channel 3')
#ax.plot(df1['Timestamp'], df1['EXG Channel 4'], label='EXG Channel 4')
ax.set_xlabel("Timestamp")
ax.set_ylabel("EXG Value")
ax.legend()
```
